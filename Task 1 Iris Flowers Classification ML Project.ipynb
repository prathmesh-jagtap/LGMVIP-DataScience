>## Visualizatoin of Data

```python
# create color dictionary
def Scatter_plot(a, b):
    colors = {'Iris-setosa':'r', 'Iris-versicolor':'g', 'Iris-virginica':'b'}
    # create a figure and axis
    fig, ax = plt.subplots()
    # plot each data-point
    for i in range(len(iris_data[a])):
        ax.scatter(iris_data[a][i], iris_data[b][i],color=colors[iris_data['Species'][i]])
    # set a title and labels
    ax.set_title('Iris Data')
    ax.set_xlabel(a)
    ax.set_ylabel(b)
Scatter_plot("SepalLengthCm", "SepalWidthCm")
Scatter_plot("PetalLengthCm", "PetalWidthCm")
```

>## Label Encoder
>Converting the categorical data into quantitative

```python
from sklearn.preprocessing import LabelEncoder
encode = LabelEncoder()
iris_data["Species"] = encode.fit_transform(iris_data["Species"])
```

```python
iris_data.head()
```

>## Selcton of Feature and Target Variable

```python
# feature variable:
X = iris_data.iloc[:, :-1]

# Target Variable:
y = iris_data.iloc[:, -1]
```

># Supervised Learning Classification Models

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import classification_report as cfr
import scikitplot as skplt
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, random_state = 42, test_size = 0.25, stratify = y)
```

># Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_fit = log_model.fit(x_train, y_train)
log_pred = log_model.predict(x_test)
y_prob = log_model.predict_proba(x_test)
print("Accuracy of log_model:", log_model.score(x_test, y_test)*100)
print("classification_report :\n", cfr(y_test, log_pred))
```

```python
skplt.metrics.plot_confusion_matrix(y_test, log_pred, figsize=(8,8), normalize=True)
skplt.metrics.plot_roc(y_test, y_prob,  figsize=(8,8))
skplt.estimators.plot_learning_curve(log_model, X, y, figsize=(8,8))
plt.show
```

># DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

tree_model = DecisionTreeClassifier(max_depth=2)
tree_fit = tree_model.fit(x_train, y_train)
tree_pred = tree_model.predict(x_test)
tree_prob = tree_model.predict_proba(x_test)
print("Accuracy of tree_model:", tree_model.score(x_test, y_test)*100)
print("classification_report :\n", cfr(y_test, tree_pred))
```

```python
skplt.metrics.plot_confusion_matrix(y_test,tree_pred, figsize=(8,8), normalize=True)
skplt.metrics.plot_roc(y_test, tree_prob,  figsize=(8,8))
skplt.estimators.plot_learning_curve(tree_model, X, y, figsize=(8,8))
plt.show()
```

```python
tree.plot_tree(tree_model)
```

># KNN Classifier 

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier(n_neighbors=2)
knn_fit = knn_model.fit(x_train, y_train)
knn_pred = knn_model.predict(x_test)
knn_prob = knn_model.predict_proba(x_test)
print("Accuracy of knn_model:", knn_model.score(x_test, y_test)*100)
print("classification_report :\n", cfr(y_test, knn_pred))
```

```python
skplt.metrics.plot_confusion_matrix(y_test,knn_pred, figsize=(8,8), normalize=True)
skplt.metrics.plot_roc(y_test, knn_prob,  figsize=(8,8))
skplt.estimators.plot_learning_curve(knn_model, X, y, figsize=(8,8))
```

># Support Vector Machine

```python
from sklearn.svm import SVC

svc_model = SVC(kernel='rbf', gamma=0.002, probability=True)
svc_fit = svc_model.fit(x_train, y_train)
svc_pred = svc_model.predict(x_test)
svc_prob = svc_model.predict_proba(x_test)
print("Accuracy of svc_model:", svc_model.score(x_test, y_test)*100)
print("classification_report :\n", cfr(y_test, svc_pred))
```

```python
skplt.metrics.plot_confusion_matrix(y_test,svc_pred, figsize=(8,8), normalize=True)
skplt.metrics.plot_roc(y_test, svc_prob,  figsize=(8,8))
skplt.estimators.plot_learning_curve(svc_model, X, y, figsize=(8,8))
```
